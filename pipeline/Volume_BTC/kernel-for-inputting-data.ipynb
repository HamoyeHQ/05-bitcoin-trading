```python
#load libraries
import numpy as np
import pandas as pd
import datetime

# read converting unix timestamp to date object and making timestamp index
df = pd.read_csv('../input/bitcoin-historical-data/bitstampUSD_1-min_data_2012-01-01_to_2020-09-14.csv',parse_dates=True, date_parser=lambda unixTime: datetime.datetime.fromtimestamp(float(unixTime)), index_col=0)

# makes a copy of 'df'. 
# 'bitcoin_df_imputation1' signifies our first approach of imputing
bitcoin_df_imputation1 = df.copy() 
 # forward fill for the 'Close' column
bitcoin_df_imputation1[['Close'].fillna(method='ffill', inplace=True)
# imputing 'Volume_(BTC)', 'Volume_(Currency)', and 'Weighted_Price' with zeros
bitcoin_df_imputation1[['Volume_(BTC)', 'Volume_(Currency)', 'Weighted_Price']] = bitcoin_df_imputation1[['Volume_(BTC)', 'Volume_(Currency)', 'Weighted_Price']].fillna(0)
# copying the values of 'Close' column to three columns of 'Open', 'High', and 'Low' that would later be used for imputation.
impute1_df = pd.DataFrame({'Open': bitcoin_df_imputation1['Close'], 'High': bitcoin_df_imputation1['Close'],'Low': bitcoin_df_imputation1['Close']})
bitcoin_df_imputation1[['Open', 'High', 'Low']] = bitcoin_df_imputation1[['Open', 'High', 'Low']].fillna(impute1_df)

# makes a copy of 'df'. 
# 'bitcoin_df_imputation2' signifies our first approach of imputing
 # drop na values
bitcoin_df_imputation2 = df.dropna()

#save datasets
bitcoin_df_imputation1.to_csv("bitcoin_df_imputation1.csv")

#save datasets
bitcoin_df_imputation2.to_csv("bitcoin_df_imputation2.csv")
                       
# Resampling
# Summing each minutes volume would suffice
                       
# By Days
BTC_data_days = bitcoin_df_imputation1['Volume_(BTC)'].resample('D').sum()

#By Weeks
BTC_data_weeks = bitcoin_df_imputation1['Volume_(BTC)'].resample('W').sum()

#By Months
BTC_data_months = bitcoin_df_imputation1['Volume_(BTC)'].resample('M').sum()
                       
#save datasets
BTC_data_days.to_csv("BTC_data_days.csv")
BTC_data_weeks.to_csv("BTC_data_week.csv")
BTC_data_months.to_csv("BTC_data_months.csv")
```
