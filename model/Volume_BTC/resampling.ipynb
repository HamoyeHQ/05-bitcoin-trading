# Resampling

In resampling I sum the volumes. This is done because the volume is described by [wikipedia](https://g.co/kgs/KWB1Bk) as 
`...the amount of a security that was traded during a given period of time. 
In the context of a single stock trading on a stock exchange, the volume is 
commonly reported as the number of shares that changed hands during a given day.`

Summing each minutes volume would suffice

By Days

```python
BTC_data_days = bitcoin_df_imputation1['Volume_(BTC)'].resample('D').sum()
```

By Weeks

```python
BTC_data_weeks = bitcoin_df_imputation1['Volume_(BTC)'].resample('W').sum()
```

By Months

```python
BTC_data_months = bitcoin_df_imputation1['Volume_(BTC)'].resample('M').sum()
```
